```python
# elephant_ml_improved.py
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Reproducibility
RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)

# sklearn imports
from sklearn.model_selection import train_test_split, cross_val_score, RandomizedSearchCV, GridSearchCV, StratifiedKFold, learning_curve
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier, IsolationForest
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_auc_score
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectFromModel
from sklearn.manifold import TSNE

# optional libs
import xgboost as xgb
import scipy.stats as stats
import time
import psutil

# try import shap (optional)
try:
    import shap
    SHAP_AVAILABLE = True
except Exception:
    SHAP_AVAILABLE = False

print("SHAP available:", SHAP_AVAILABLE)

# -------------------------
# Load dataset
# -------------------------
df = pd.read_csv('20231225_dfall_obs_data_and_spectral_features_revision1_n469.csv')
print("Dataset Shape:", df.shape)
print("Memory usage (MB):", df.memory_usage(deep=True).sum() / 1024**2)

# -------------------------
# Feature and target setup
# -------------------------
feature_columns = [f'V{i}' for i in range(1, 75)] + \
                [f'Fw{i}' for i in range(1, 5)] + [f'Mw{i}' for i in range(1, 5)] + \
                [f'F{i}' for i in range(1, 5)] + [f'M{i}' for i in range(1, 5)] + \
                ['sprsMed', 'sprsMbw', 'sprsEqbw', 'sprsMc']

available_features = [col for col in feature_columns if col in df.columns]
print(f"Available features for modeling: {len(available_features)}")

targets = {
    'age': 'CallerBinAge',  # Adult vs Juvenile
    'sex': 'CallerSex',     # Male vs Female
    'emotion': 'Context2'    # Context as proxy for emotion/behavior
}

print("\n=== TARGET VARIABLE DISTRIBUTIONS ===")
for tname, tcol in targets.items():
    if tcol in df.columns:
        print(f"\n{tname.upper()} distribution:")
        print(df[tcol].value_counts())
    else:
        print(f"Warning: {tcol} not found in dataset")

# -------------------------
# Basic cleaning
# -------------------------
df_clean = df.copy()

# Fill feature NAs with mean
df_clean[available_features] = df_clean[available_features].fillna(df_clean[available_features].mean())

# Remove rows with missing targets
for target_col in targets.values():
    if target_col in df_clean.columns:
        df_clean = df_clean[df_clean[target_col].notna()]

print(f"Dataset size after cleaning: {df_clean.shape}")

# -------------------------
# Outlier detection (IsolationForest)
# -------------------------
print("\nRunning outlier detection (IsolationForest)...")
if len(df_clean) > 50 and len(available_features) > 0:
    iso = IsolationForest(contamination=0.02, random_state=RANDOM_STATE)
    try:
        iso_preds = iso.fit_predict(df_clean[available_features])
        n_outliers = np.sum(iso_preds == -1)
        mask = iso_preds != -1
        if n_outliers > 0:
            print(f"Detected {n_outliers} outliers. Removing them.")
            df_clean = df_clean.loc[mask].reset_index(drop=True)
        else:
            print("No outliers detected.")
    except Exception as e:
        print("IsolationForest failed:", e)
else:
    print("Not enough data/features to run outlier detection.")

print(f"Dataset size after outlier removal: {df_clean.shape}")

# -------------------------
# Prepare X and label-encode targets
# -------------------------
X_raw = df_clean[available_features].copy()
y_encoded = {}
label_encoders = {}

for tname, tcol in targets.items():
    if tcol in df_clean.columns:
        le = LabelEncoder()
        y_encoded[tname] = le.fit_transform(df_clean[tcol])
        label_encoders[tname] = le
        print(f"{tname} classes: {list(le.classes_)}")

# -------------------------
# Feature scaling (fit later on selected features)
# -------------------------
scaler = StandardScaler()

# -------------------------
# Automated feature selection (SelectFromModel using RandomForest on a baseline target)
# -------------------------
print("\nRunning automated feature selection (SelectFromModel using RandomForest)...")
selected_feature_mask = None
if len(available_features) > 0 and len(y_encoded) > 0:
    # choose a baseline target for selection: prefer 'age' if present
    baseline_target = 'age' if 'age' in y_encoded else list(y_encoded.keys())[0]
    try:
        rf_fs = RandomForestClassifier(n_estimators=200, random_state=RANDOM_STATE, n_jobs=-1)
        rf_fs.fit(X_raw.fillna(0), y_encoded[baseline_target])
        selector = SelectFromModel(rf_fs, prefit=True, threshold='median')
        selected_feature_mask = selector.get_support()
        selected_features = [f for f, keep in zip(available_features, selected_feature_mask) if keep]
        print(f"Selected {len(selected_features)} features out of {len(available_features)} using target '{baseline_target}'.")
    except Exception as e:
        print("Feature selection failed, falling back to all features:", e)
        selected_features = available_features.copy()
else:
    selected_features = available_features.copy()

# Ensure we have at least a few features
if len(selected_features) < 5:
    selected_features = available_features.copy()
    print("Too few features selected; using all available features.")

# prepare final X
X = df_clean[selected_features].copy()
print("Final feature set size:", X.shape[1])

# scale
X_scaled = scaler.fit_transform(X)

# -------------------------
# Train-test split for each target (stratified)
# -------------------------
X_train, X_test, y_train_dict, y_test_dict = {}, {}, {}, {}
for tname in y_encoded.keys():
    X_train[tname], X_test[tname], y_train_dict[tname], y_test_dict[tname] = train_test_split(
        X_scaled, y_encoded[tname], test_size=0.3, random_state=RANDOM_STATE, stratify=y_encoded[tname]
    )

# -------------------------
# Models & hyperparameter tuning (RandomizedSearch for heavier models)
# -------------------------
print("\nSetting up models and hyperparameter tuning...")

# Base estimators with default small budgets
base_estimators = {
    'Random Forest': RandomForestClassifier(n_estimators=200, random_state=RANDOM_STATE, n_jobs=-1),
    'XGBoost': xgb.XGBClassifier(n_estimators=200, use_label_encoder=False, eval_metric='logloss', random_state=RANDOM_STATE, n_jobs=1),
    'SVM': SVC(probability=True, random_state=RANDOM_STATE),
    'Gradient Boosting': GradientBoostingClassifier(random_state=RANDOM_STATE),
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=RANDOM_STATE)
}

# Parameter distributions for RandomizedSearchCV
rf_param_dist = {
    'n_estimators': [100, 200, 400],
    'max_depth': [None, 5, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

xgb_param_dist = {
    'n_estimators': [100, 200, 400],
    'max_depth': [3, 5, 7, 10],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'subsample': [0.6, 0.8, 1.0]
}

svc_param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['rbf', 'linear'],
    'gamma': ['scale', 'auto']
}

logreg_param_grid = {
    'C': [0.01, 0.1, 1, 10],
    'penalty': ['l2']
}

# a helper function to tune a model (uses RandomizedSearch for RF/XGB, Grid for small grids)
def tune_model(name, estimator, X_tr, y_tr, n_iter=20):
    cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=RANDOM_STATE)
    if name == 'Random Forest':
        search = RandomizedSearchCV(estimator, rf_param_dist, n_iter=n_iter, cv=cv, n_jobs=-1, random_state=RANDOM_STATE, scoring='accuracy')
    elif name == 'XGBoost':
        search = RandomizedSearchCV(estimator, xgb_param_dist, n_iter=n_iter, cv=cv, n_jobs=1, random_state=RANDOM_STATE, scoring='accuracy')
    elif name == 'SVM':
        search = GridSearchCV(estimator, svc_param_grid, cv=cv, n_jobs=-1, scoring='accuracy')
    elif name == 'Logistic Regression':
        search = GridSearchCV(estimator, logreg_param_grid, cv=cv, n_jobs=-1, scoring='accuracy')
    else:
        # small default CV
        search = RandomizedSearchCV(estimator, {}, n_iter=1, cv=cv, n_jobs=-1)
    
    try:
        start = time.time()
        search.fit(X_tr, y_tr)
        duration = time.time() - start
        print(f"Tuning {name} done in {duration:.1f}s. Best score: {search.best_score_:.3f}")
        return search.best_estimator_, search.best_params_
    except Exception as e:
        print(f"Tuning for {name} failed or skipped (using default). Error: {e}")
        estimator.fit(X_tr, y_tr)
        return estimator, {}

# -------------------------
# Evaluation function (uses stratified CV)
# -------------------------
def evaluate_model(model, X_tr, X_te, y_tr, y_te, target_name, cv_folds=5):
    # train
    start = time.time()
    model.fit(X_tr, y_tr)
    train_time = time.time() - start
    
    y_pred = model.predict(X_te)
    y_proba = model.predict_proba(X_te) if hasattr(model, 'predict_proba') else None
    
    acc = accuracy_score(y_te, y_pred)
    cv = StratifiedKFold(n_splits=cv_folds, shuffle=True, random_state=RANDOM_STATE)
    try:
        cv_scores = cross_val_score(model, X_tr, y_tr, cv=cv, scoring='accuracy', n_jobs=-1)
        cv_mean, cv_std = cv_scores.mean(), cv_scores.std()
    except Exception:
        cv_mean, cv_std = np.nan, np.nan
    
    roc_auc = None
    if y_proba is not None and len(np.unique(y_te)) == 2:
        try:
            roc_auc = roc_auc_score(y_te, y_proba[:, 1])
        except Exception:
            roc_auc = None
    
    return {
        'model': model,
        'accuracy': acc,
        'cv_mean': cv_mean,
        'cv_std': cv_std,
        'roc_auc': roc_auc,
        'y_pred': y_pred,
        'y_test': y_te,
        'train_time_s': train_time
    }

# -------------------------
# Train, tune and evaluate models for each target
# -------------------------
results = {}
accuracy_data = []

for tname in y_encoded.keys():
    print("\n" + "="*60)
    print(f"PROCESSING TARGET: {tname.upper()}")
    print("="*60)
    
    X_tr = X_train[tname]
    X_te = X_test[tname]
    y_tr = y_train_dict[tname]
    y_te = y_test_dict[tname]
    
    # Tune select models (RF, XGB, SVM, LogReg). Keep GB default (could tune similarly)
    tuned_estimators = {}
    for mname, base in base_estimators.items():
        print(f"\nTuning/training: {mname} for target {tname} ...")
        est_copy = base  # note: sklearn estimators are usually stateless until fit
        best_est, best_params = tune_model(mname, est_copy, X_tr, y_tr, n_iter=20)
        tuned_estimators[mname] = best_est
    
    # Add a soft VotingClassifier ensemble of top 3 if possible
    ensemble = None
    try:
        voting_estimators = []
        prefer_order = ['Random Forest', 'XGBoost', 'Gradient Boosting', 'SVM', 'Logistic Regression']
        added = 0
        for k in prefer_order:
            if k in tuned_estimators and added < 3:
                voting_estimators.append((k.replace(" ", "_"), tuned_estimators[k]))
                added += 1
        if len(voting_estimators) >= 2:
            ensemble = VotingClassifier(estimators=voting_estimators, voting='soft', n_jobs=-1)
            tuned_estimators['Ensemble_Voting'] = ensemble
            print("Ensemble (Voting) created with:", [n for n, _ in voting_estimators])
    except Exception as e:
        print("Failed to create ensemble:", e)
    
    # Evaluate each tuned estimator for this target
    target_results = {}
    for mname, estimator in tuned_estimators.items():
        print(f"Evaluating {mname} ...")
        try:
            res = evaluate_model(estimator, X_tr, X_te, y_tr, y_te, tname, cv_folds=5)
            target_results[mname] = res
            accuracy_data.append({'Target': tname, 'Model': mname, 'Accuracy': res['accuracy']})
            print(f"{mname} - Acc: {res['accuracy']:.3f}, CV: {res['cv_mean']:.3f}  {res['cv_std']:.3f}, Time(s): {res['train_time_s']:.1f}")
            if res['roc_auc'] is not None:
                print(f"ROC AUC: {res['roc_auc']:.3f}")
        except Exception as e:
            print(f"Error evaluating {mname} for {tname}: {e}")
    
    results[tname] = target_results

# -------------------------
# Visualization: Model comparison barplots (accuracy)
# -------------------------
print("\nPlotting model comparison charts...")
accuracy_df = pd.DataFrame(accuracy_data)
if not accuracy_df.empty:
    plt.figure(figsize=(10, 6))
    sns.barplot(data=accuracy_df, x='Target', y='Accuracy', hue='Model')
    plt.title('Model Comparison Across Targets')
    plt.xticks(rotation=45)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

# -------------------------
# Feature importance (from Random Forest) + SHAP (optional)
# -------------------------
if 'age' in results and 'Random Forest' in results['age']:
    try:
        best_rf = results['age']['Random Forest']['model']
        fi = best_rf.feature_importances_
        fi_df = pd.DataFrame({'feature': selected_features, 'importance': fi}).sort_values('importance', ascending=False).head(20)
        plt.figure(figsize=(8, 6))
        sns.barplot(data=fi_df, x='importance', y='feature')
        plt.title('Top features (Random Forest - age)')
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print("Could not plot RF feature importance:", e)

# SHAP explainability (optional)
if SHAP_AVAILABLE and 'age' in results and 'Random Forest' in results['age']:
    try:
        print("\nRunning SHAP explainability (this may take a bit)...")
        explainer = shap.TreeExplainer(results['age']['Random Forest']['model'])
        # Use a sample to speed up
        sample_X = X.sample(n=min(200, len(X)), random_state=RANDOM_STATE)
        shap_values = explainer.shap_values(sample_X)
        # summary plot (for multiclass shap_values may be list)
        shap.summary_plot(shap_values, sample_X, feature_names=selected_features, show=True)
    except Exception as e:
        print("SHAP plotting failed:", e)
else:
    if not SHAP_AVAILABLE:
        print("SHAP not installed - skipping SHAP explainability.")

# -------------------------
# PCA & t-SNE visualization
# -------------------------
if 'age' in y_encoded:
    print("\nPCA + t-SNE visualization for 'age' target")
    try:
        pca = PCA(n_components=2, random_state=RANDOM_STATE)
        X_pca = pca.fit_transform(X_scaled)
        plt.figure(figsize=(8, 6))
        scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_encoded['age'], alpha=0.6)
        plt.title(f'PCA - Age Classification (PC1 {pca.explained_variance_ratio_[0]:.2%}, PC2 {pca.explained_variance_ratio_[1]:.2%})')
        plt.xlabel('PC1'); plt.ylabel('PC2')
        plt.legend(handles=scatter.legend_elements()[0], labels=list(label_encoders['age'].classes_))
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print("PCA plotting failed:", e)
    
    try:
        tsne = TSNE(n_components=2, random_state=RANDOM_STATE, perplexity=30, n_iter=1000)
        X_tsne = tsne.fit_transform(X_scaled)
        plt.figure(figsize=(8, 6))
        sns.scatterplot(x=X_tsne[:, 0], y=X_tsne[:, 1], hue=y_encoded['age'], palette='viridis', alpha=0.8)
        plt.title('t-SNE projection - Age')
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print("t-SNE failed (may be slow):", e)

# -------------------------
# Detailed classification reports + confusion matrices
# -------------------------
print("\n" + "="*60)
print("DETAILED CLASSIFICATION REPORTS")
print("="*60)

for tname, tres in results.items():
    if len(tres) == 0:
        continue
    # pick best by accuracy
    best_name = max(tres.keys(), key=lambda x: tres[x]['accuracy'])
    best_res = tres[best_name]
    print(f"\nTARGET: {tname.upper()}  |  BEST MODEL: {best_name}")
    print(f"Accuracy: {best_res['accuracy']:.3f}")
    if not np.isnan(best_res['cv_mean']):
        print(f"5-fold CV: {best_res['cv_mean']:.3f}  {best_res['cv_std']:.3f}")
    if best_res['roc_auc'] is not None:
        print(f"ROC AUC: {best_res['roc_auc']:.3f}")
    
    # classification report
    try:
        print("\nClassification Report:")
        print(classification_report(best_res['y_test'], best_res['y_pred'], target_names=label_encoders[tname].classes_))
    except Exception as e:
        print("Failed to produce classification report:", e)
    
    # confusion matrix plot
    try:
        cm = confusion_matrix(best_res['y_test'], best_res['y_pred'])
        plt.figure(figsize=(5, 4))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                    xticklabels=label_encoders[tname].classes_,
                    yticklabels=label_encoders[tname].classes_)
        plt.title(f'Confusion Matrix - {tname.upper()} ({best_name})')
        plt.ylabel('True'); plt.xlabel('Predicted')
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print("Failed to plot confusion matrix:", e)

# -------------------------
# Learning curve for best age model (if exists)
# -------------------------
if 'age' in results and len(results['age']) > 0:
    try:
        best_age_model_name = max(results['age'].keys(), key=lambda x: results['age'][x]['accuracy'])
        best_age_model = results['age'][best_age_model_name]['model']
        print(f"\nPlotting learning curve for Age - {best_age_model_name}")
        
        plt.figure(figsize=(10, 6))
        plt.title(f"Learning Curve - Age Classification ({best_age_model_name})")
        plt.xlabel("Training examples")
        plt.ylabel("Score")
        cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
        train_sizes, train_scores, test_scores = learning_curve(best_age_model, X_train['age'], y_train_dict['age'], cv=cv, n_jobs=-1, train_sizes=np.linspace(.1, 1.0, 5))
        train_scores_mean = np.mean(train_scores, axis=1)
        train_scores_std  = np.std(train_scores, axis=1)
        test_scores_mean  = np.mean(test_scores, axis=1)
        test_scores_std   = np.std(test_scores, axis=1)
        plt.grid()
        plt.fill_between(train_sizes, train_scores_mean - train_scores_std, train_scores_mean + train_scores_std, alpha=0.1)
        plt.fill_between(train_sizes, test_scores_mean - test_scores_std, test_scores_mean + test_scores_std, alpha=0.1)
        plt.plot(train_sizes, train_scores_mean, 'o-', label='Training score')
        plt.plot(train_sizes, test_scores_mean, 'o-', label='Cross-validation score')
        plt.legend(loc='best')
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print("Learning curve plotting failed:", e)

# -------------------------
# Summary statistics
# -------------------------
print("\n" + "="*60)
print("SUMMARY STATISTICS")
print("="*60)
if len(accuracy_data) > 0:
    accuracy_summary = pd.DataFrame(accuracy_data).pivot_table(values='Accuracy', index='Model', columns='Target', aggfunc='mean')
    print("\nAccuracy Summary by Model and Target:")
    print(accuracy_summary.round(3))
    print("\nOVERALL BEST MODELS:")
    for tname in results.keys():
        if results[tname]:
            best_m = max(results[tname].items(), key=lambda x: x[1]['accuracy'])
            print(f"{tname.upper()}: {best_m[0]} (Accuracy: {best_m[1]['accuracy']:.3f})")
else:
    print("No accuracy data collected.")

print("\nCONCLUSIONS AND RECOMMENDATIONS (auto)")
print("""
1. Check class balance; if heavily imbalanced consider class weights or resampling (SMOTE).
2. Use ensemble (soft voting)  often gives robust gains.
3. If SHAP is available, use it to validate biologically meaningful features.
4. Consider temporal/deep models if you have sequential audio (RNN/CNN on spectrograms).
5. If compute/memory is a concern, reduce selected features or use feature selection thresholding.
""")

print("\nDone. LMK if you want this trimmed down, or want me to enable saving models, or to prepare a Streamlit dashboard for interactive exploration.")
```

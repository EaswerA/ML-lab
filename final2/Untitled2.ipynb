```python
# fixed_emotion_pipeline.py
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split, cross_val_score, learning_curve
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_auc_score
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest, f_classif

# optional libs
import xgboost as xgb

# Try import SMOTE (imblearn); if not available, we'll fallback to class_weight or no resampling
try:
    from imblearn.over_sampling import SMOTE
    SMOTE_AVAILABLE = True
except Exception:
    SMOTE_AVAILABLE = False

print("SMOTE available:", SMOTE_AVAILABLE)

# ---------------------------
# Load the new cepstral dataset
# ---------------------------
df = pd.read_csv("20240106_dfall_obs_data_and_cepstral_features_revision1_n469.csv")
print("Dataset Shape:", df.shape)
print("Columns count:", len(df.columns))

# ---------------------------
# Feature discovery (cepstral/ spectral-like columns)
# ---------------------------
# adapt to columns that begin with typical feature prefixes
feature_columns = [col for col in df.columns if col.startswith('C') or col.startswith('V') 
                   or col.startswith('F') or col.startswith('M') or col.startswith('sprs')]

# Now keep only those present
available_features = [col for col in feature_columns if col in df.columns]
print(f"Initial available features (by prefix): {len(available_features)}")

# IMPORTANT: restrict to numeric columns only (this fixes the TypeError you saw)
numeric_features = df[available_features].select_dtypes(include=[np.number]).columns.tolist()
non_numeric = set(available_features) - set(numeric_features)
if non_numeric:
    print("Dropped non-numeric feature columns (cannot take mean):", list(non_numeric))
print(f"Numeric features available for modeling: {len(numeric_features)}")

if len(numeric_features) == 0:
    raise RuntimeError("No numeric features found among detected prefixes. Check column names.")

# ---------------------------
# Targets
# ---------------------------
targets = {
    'age': 'CallerBinAge',
    'sex': 'CallerSex',
    'emotion': 'Context2'
}

print("\n=== TARGET DISTRIBUTIONS ===")
for name, col in targets.items():
    if col in df.columns:
        print(f"\n{name.upper()} distribution:")
        print(df[col].value_counts())
    else:
        print(f"Warning: {col} not in dataframe")

# ---------------------------
# Basic cleaning / fill numeric NAs with numeric column means
# ---------------------------
df_clean = df.copy()
# convert numeric_features to numeric coercing errors (if any stray strings)
for col in numeric_features:
    df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')

# fill NAs in numeric features with column means
df_clean[numeric_features] = df_clean[numeric_features].fillna(df_clean[numeric_features].mean())

# drop rows with missing targets
for tcol in targets.values():
    if tcol in df_clean.columns:
        df_clean = df_clean[df_clean[tcol].notna()]

print(f"Dataset shape after cleaning targets & numeric NA-fill: {df_clean.shape}")

# ---------------------------
# Prepare X and y for emotion
# ---------------------------
X = df_clean[numeric_features].copy()

# ensure 'emotion' exists
if targets['emotion'] not in df_clean.columns:
    raise RuntimeError("Emotion target column (Context2) not found in dataset.")

le = LabelEncoder()
y_emotion = le.fit_transform(df_clean[targets['emotion']])
print("Emotion classes:", list(le.classes_))
num_classes = len(le.classes_)

# ---------------------------
# Scale features
# ---------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ---------------------------
# Feature selection - SelectKBest for emotion (ANOVA F)
# ---------------------------
k = min(40, X.shape[1])  # keep up to top 40 features or fewer if not available
selector = SelectKBest(score_func=f_classif, k=k)
# SelectKBest requires y with more than one class
if len(np.unique(y_emotion)) < 2:
    raise RuntimeError("Less than 2 emotion classes found — cannot do classification.")

X_selected = selector.fit_transform(X_scaled, y_emotion)
selected_idx = selector.get_support(indices=True)
selected_features = [numeric_features[i] for i in selected_idx]
print(f"Selected top-{len(selected_features)} features for emotion. Example: {selected_features[:10]}")

# ---------------------------
# Train-test split and optional SMOTE balancing
# ---------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_selected, y_emotion, test_size=0.30, random_state=42, stratify=y_emotion
)
print("Train/test sizes:", X_train.shape, X_test.shape)

if SMOTE_AVAILABLE:
    smote = SMOTE(random_state=42)
    X_train_res, y_train_res = smote.fit_resample(X_train, y_train)
    print("After SMOTE class counts:", np.bincount(y_train_res))
else:
    # fallback: check imbalance and optionally use class weights in models
    X_train_res, y_train_res = X_train, y_train
    print("SMOTE not available — using original training set (consider installing imblearn).")
    print("Train class counts:", np.bincount(y_train_res))

# ---------------------------
# Models (with class_weight where helpful)
# ---------------------------
models = {
    'Random Forest': RandomForestClassifier(n_estimators=200, random_state=42, class_weight='balanced'),
    'XGBoost': xgb.XGBClassifier(n_estimators=200, use_label_encoder=False, eval_metric='mlogloss', random_state=42),
    'SVM': SVC(probability=True, kernel='rbf', random_state=42, class_weight='balanced'),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42),
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42, class_weight='balanced' if not SMOTE_AVAILABLE else None)
}

# Ensemble voting (soft) — will use the above models; some models may not support predict_proba in scikit versions
try:
    voting = VotingClassifier(estimators=[(name, model) for name, model in models.items()], voting='soft')
    models['EnsembleVoting'] = voting
except Exception as e:
    print("Could not create VotingClassifier:", e)

# ---------------------------
# Evaluation utils
# ---------------------------
def evaluate_model(model, X_tr, X_te, y_tr, y_te):
    model.fit(X_tr, y_tr)
    y_pred = model.predict(X_te)
    y_proba = model.predict_proba(X_te) if hasattr(model, 'predict_proba') else None

    acc = accuracy_score(y_te, y_pred)
    try:
        cv_scores = cross_val_score(model, X_tr, y_tr, cv=5, scoring='accuracy', n_jobs=-1)
        cv_mean, cv_std = cv_scores.mean(), cv_scores.std()
    except Exception:
        cv_mean, cv_std = np.nan, np.nan

    roc_auc = None
    if y_proba is not None and num_classes == 2:
        try:
            roc_auc = roc_auc_score(y_te, y_proba[:, 1])
        except Exception:
            roc_auc = None
    elif y_proba is not None and num_classes > 2:
        try:
            roc_auc = roc_auc_score(y_te, y_proba, multi_class='ovr')
        except Exception:
            roc_auc = None

    return {
        'model': model,
        'accuracy': acc,
        'cv_mean': cv_mean,
        'cv_std': cv_std,
        'roc_auc': roc_auc,
        'y_pred': y_pred,
        'y_test': y_te
    }

# ---------------------------
# Train & evaluate each model
# ---------------------------
results = {}
for name, model in models.items():
    try:
        print(f"\nTraining -> {name}")
        res = evaluate_model(model, X_train_res, X_test, y_train_res, y_test)
        results[name] = res
        print(f"{name}: Acc={res['accuracy']:.3f}, CV={res['cv_mean']:.3f}, ROC_AUC={res['roc_auc']}")
    except Exception as e:
        print(f"Failed training {name}: {e}")

# ---------------------------
# Plot accuracies
# ---------------------------
accs = [(n, r['accuracy']) for n, r in results.items()]
if accs:
    names, vals = zip(*accs)
    plt.figure(figsize=(10,5))
    sns.barplot(x=list(names), y=list(vals))
    plt.title("Emotion classification accuracy (models)")
    plt.ylabel("Accuracy")
    plt.xticks(rotation=45)
    for i, v in enumerate(vals):
        plt.text(i, v + 0.01, f"{v:.3f}", ha='center')
    plt.tight_layout()
    plt.show()

# ---------------------------
# Best model classification report + confusion matrix
# ---------------------------
if results:
    best_name = max(results.keys(), key=lambda k: results[k]['accuracy'])
    best = results[best_name]
    print(f"\nBEST MODEL: {best_name}  (Accuracy: {best['accuracy']:.3f})")
    print("\nClassification Report:")
    print(classification_report(best['y_test'], best['y_pred'], target_names=le.classes_))

    cm = confusion_matrix(best['y_test'], best['y_pred'])
    plt.figure(figsize=(7,6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=le.classes_, yticklabels=le.classes_)
    plt.title(f"Confusion Matrix - {best_name}")
    plt.xlabel("Predicted"); plt.ylabel("True")
    plt.show()

    # Feature importance if available
    if hasattr(best['model'], 'feature_importances_'):
        importances = best['model'].feature_importances_
        imp_df = pd.DataFrame({'feature': selected_features, 'importance': importances})
        imp_df = imp_df.sort_values('importance', ascending=False).head(20)
        plt.figure(figsize=(8,6))
        sns.barplot(x='importance', y='feature', data=imp_df)
        plt.title("Top features (by importance)")
        plt.tight_layout()
        plt.show()

# ---------------------------
# PCA visualization (2D) of selected features
# ---------------------------
try:
    pca = PCA(n_components=2, random_state=42)
    X_pca = pca.fit_transform(X_selected)
    plt.figure(figsize=(8,6))
    scatter = plt.scatter(X_pca[:,0], X_pca[:,1], c=y_emotion, cmap='viridis', alpha=0.7)
    plt.title("PCA - Emotion (selected features)")
    plt.xlabel("PC1"); plt.ylabel("PC2")
    plt.legend(handles=scatter.legend_elements()[0], labels=list(le.classes_))
    plt.tight_layout()
    plt.show()
except Exception as e:
    print("PCA failed:", e)

# ---------------------------
# Learning curve for the best model (if exists)
# ---------------------------
try:
    if 'best' in locals():
        estimator = best['model']
        plt.figure(figsize=(10,6))
        train_sizes, train_scores, test_scores = learning_curve(
            estimator, X_train_res, y_train_res, cv=5, n_jobs=-1,
            train_sizes=np.linspace(0.1,1.0,5)
        )
        plt.plot(train_sizes, np.mean(train_scores, axis=1), 'o-', label='Train')
        plt.plot(train_sizes, np.mean(test_scores, axis=1), 'o-', label='CV')
        plt.legend(); plt.xlabel("Training sizes"); plt.ylabel("Score")
        plt.title(f"Learning curve - {best_name}")
        plt.grid(); plt.tight_layout(); plt.show()
except Exception as e:
    print("Learning curve failed:", e)

print("\nDone.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_auc_score
from sklearn.decomposition import PCA
import xgboost as xgb
import warnings
warnings.filterwarnings('ignore')

# Load the dataset
df = pd.read_csv('20231225_dfall_obs_data_and_spectral_features_revision1_n469.csv')

# Display basic info about the dataset
print("Dataset Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nFirst few rows:")
print(df.head())

# Data Preprocessing and Feature Selection
print("\n=== DATA PREPROCESSING ===")

# Select relevant features for classification
# Using spectral features (V1-V74), frequency features (Fw1-Fw4, Mw1-Mw4, F1-F4, M1-M4), and other acoustic features
feature_columns = [f'V{i}' for i in range(1, 75)] + \
                [f'Fw{i}' for i in range(1, 5)] + [f'Mw{i}' for i in range(1, 5)] + \
                [f'F{i}' for i in range(1, 5)] + [f'M{i}' for i in range(1, 5)] + \
                ['sprsMed', 'sprsMbw', 'sprsEqbw', 'sprsMc']

# Check which features exist in our dataset
available_features = [col for col in feature_columns if col in df.columns]
print(f"Available features for modeling: {len(available_features)}")

# Target variables
targets = {
    'age': 'CallerBinAge',  # Adult vs Juvenile
    'sex': 'CallerSex',     # Male vs Female
    'emotion': 'Context2'    # Context as proxy for emotion/behavior
}

# Check target variable distributions
print("\n=== TARGET VARIABLE DISTRIBUTIONS ===")
for target_name, target_col in targets.items():
    if target_col in df.columns:
        print(f"\n{target_name.upper()} distribution:")
        print(df[target_col].value_counts())
    else:
        print(f"Warning: {target_col} not found in dataset")

# Data cleaning
df_clean = df.copy()

# Handle missing values in features
df_clean[available_features] = df_clean[available_features].fillna(df_clean[available_features].mean())

# Remove rows with missing target values
for target_col in targets.values():
    if target_col in df_clean.columns:
        df_clean = df_clean[df_clean[target_col].notna()]

print(f"Dataset size after cleaning: {df_clean.shape}")

# Prepare features and targets
X = df_clean[available_features]
print(f"Feature matrix shape: {X.shape}")

# Encoding target variables
label_encoders = {}
y_encoded = {}

for target_name, target_col in targets.items():
    if target_col in df_clean.columns:
        le = LabelEncoder()
        y_encoded[target_name] = le.fit_transform(df_clean[target_col])
        label_encoders[target_name] = le
        print(f"\n{target_name} classes: {list(le.classes_)}")

# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data
X_train, X_test, y_train_dict, y_test_dict = {}, {}, {}, {}

for target_name in y_encoded.keys():
    X_train[target_name], X_test[target_name], y_train_dict[target_name], y_test_dict[target_name] = train_test_split(
        X_scaled, y_encoded[target_name], test_size=0.3, random_state=42, stratify=y_encoded[target_name]
    )

# Model Definitions
models = {
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'XGBoost': xgb.XGBClassifier(random_state=42),
    'SVM': SVC(probability=True, random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42),
    'Logistic Regression': LogisticRegression(random_state=42)
}

# Evaluation function
def evaluate_model(model, X_train, X_test, y_train, y_test, target_name):
    """Evaluate model performance"""
    # Train model
    model.fit(X_train, y_train)
    
    # Predictions
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test) if hasattr(model, 'predict_proba') else None
    
    # Metrics
    accuracy = accuracy_score(y_test, y_pred)
    cv_scores = cross_val_score(model, X_train, y_train, cv=5)
    
    # ROC AUC (if binary classification and probabilities available)
    roc_auc = None
    if y_pred_proba is not None and len(np.unique(y_test)) == 2:
        roc_auc = roc_auc_score(y_test, y_pred_proba[:, 1])
    
    return {
        'model': model,
        'accuracy': accuracy,
        'cv_mean': cv_scores.mean(),
        'cv_std': cv_scores.std(),
        'roc_auc': roc_auc,
        'y_pred': y_pred,
        'y_test': y_test
    }

# Train and evaluate models for each target
print("\n" + "="*50)
print("MODEL TRAINING AND EVALUATION")
print("="*50)

results = {}

for target_name in y_encoded.keys():
    print(f"\n=== CLASSIFYING {target_name.upper()} ===")
    
    target_results = {}
    
    for model_name, model in models.items():
        print(f"Training {model_name}...")
        
        try:
            result = evaluate_model(
                model, 
                X_train[target_name], 
                X_test[target_name], 
                y_train_dict[target_name], 
                y_test_dict[target_name],
                target_name
            )
            
            target_results[model_name] = result
            
            print(f"{model_name} - Accuracy: {result['accuracy']:.3f}, CV Score: {result['cv_mean']:.3f} ± {result['cv_std']:.3f}")
            if result['roc_auc']:
                print(f"ROC AUC: {result['roc_auc']:.3f}")
                
        except Exception as e:
            print(f"Error training {model_name} for {target_name}: {e}")
    
    results[target_name] = target_results

# Visualization and Results Analysis
print("\n" + "="*50)
print("RESULTS VISUALIZATION AND ANALYSIS")
print("="*50)

# 1. Model Performance Comparison
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
axes = axes.flatten()

# Accuracy comparison
accuracy_data = []
for i, target_name in enumerate(results.keys()):
    model_names = list(results[target_name].keys())
    accuracies = [results[target_name][model]['accuracy'] for model in model_names]
    
    axes[i].bar(model_names, accuracies, color=['skyblue', 'lightcoral', 'lightgreen', 'gold', 'lightpink'])
    axes[i].set_title(f'{target_name.upper()} Classification Accuracy')
    axes[i].set_ylabel('Accuracy')
    axes[i].tick_params(axis='x', rotation=45)
    
    # Add value labels on bars
    for j, v in enumerate(accuracies):
        axes[i].text(j, v + 0.01, f'{v:.3f}', ha='center', va='bottom')
    
    # Store for summary
    for model, acc in zip(model_names, accuracies):
        accuracy_data.append({'Target': target_name, 'Model': model, 'Accuracy': acc})

# Feature Importance for best model (using Random Forest as example)
if 'age' in results and 'Random Forest' in results['age']:
    best_model_age = results['age']['Random Forest']['model']
    feature_importance = best_model_age.feature_importances_
    
    # Get top 15 features
    importance_df = pd.DataFrame({
        'feature': available_features,
        'importance': feature_importance
    }).sort_values('importance', ascending=False).head(15)
    
    axes[2].barh(importance_df['feature'], importance_df['importance'])
    axes[2].set_title('Top 15 Features for Age Classification\n(Random Forest)')
    axes[2].set_xlabel('Feature Importance')

# PCA Visualization for age classification
if 'age' in y_encoded:
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X_scaled)
    
    scatter = axes[3].scatter(X_pca[:, 0], X_pca[:, 1], 
                            c=y_encoded['age'], 
                            cmap='viridis', 
                            alpha=0.6)
    axes[3].set_title('PCA - Age Classification')
    axes[3].set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.2%} variance)')
    axes[3].set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.2%} variance)')
    axes[3].legend(handles=scatter.legend_elements()[0], 
                 labels=list(label_encoders['age'].classes_))

plt.tight_layout()
plt.show()

# 2. Detailed Classification Reports
print("\n" + "="*50)
print("DETAILED CLASSIFICATION REPORTS")
print("="*50)

for target_name in results.keys():
    print(f"\n=== BEST MODEL FOR {target_name.upper()} ===")
    
    # Find best model for this target
    best_model_name = max(results[target_name].keys(), 
                        key=lambda x: results[target_name][x]['accuracy'])
    best_result = results[target_name][best_model_name]
    
    print(f"Best Model: {best_model_name}")
    print(f"Accuracy: {best_result['accuracy']:.3f}")
    print(f"5-fold CV Score: {best_result['cv_mean']:.3f} ± {best_result['cv_std']:.3f}")
    
    if best_result['roc_auc']:
        print(f"ROC AUC: {best_result['roc_auc']:.3f}")
    
    # Classification report
    print(f"\nClassification Report for {target_name}:")
    print(classification_report(best_result['y_test'], best_result['y_pred'], 
                              target_names=label_encoders[target_name].classes_))
    
    # Confusion Matrix
    plt.figure(figsize=(6, 5))
    cm = confusion_matrix(best_result['y_test'], best_result['y_pred'])
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=label_encoders[target_name].classes_,
                yticklabels=label_encoders[target_name].classes_)
    plt.title(f'Confusion Matrix - {target_name.upper()}\n({best_model_name})')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

# 3. Summary Statistics
print("\n" + "="*50)
print("SUMMARY STATISTICS")
print("="*50)

accuracy_df = pd.DataFrame(accuracy_data)
summary = accuracy_df.pivot_table(values='Accuracy', index='Model', columns='Target', aggfunc='mean')

print("\nAccuracy Summary by Model and Target:")
print(summary.round(3))

# Overall best models
print("\nOVERALL BEST MODELS:")
for target_name in results.keys():
    best_model = max(results[target_name].items(), key=lambda x: x[1]['accuracy'])
    print(f"{target_name.upper()}: {best_model[0]} (Accuracy: {best_model[1]['accuracy']:.3f})")

# 4. Feature Analysis
print("\n" + "="*50)
print("FEATURE ANALYSIS")
print("="*50)

# Correlation analysis
correlation_matrix = pd.DataFrame(X_scaled, columns=available_features).corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix.iloc[:20, :20], cmap='coolwarm', center=0)
plt.title('Feature Correlation Matrix (First 20 Features)')
plt.tight_layout()
plt.show()

# Statistical summary of features
feature_stats = pd.DataFrame({
    'mean': X.mean(),
    'std': X.std(),
    'min': X.min(),
    'max': X.max()
})

print("\nFeature Statistics (Top 10 features by variance):")
print(feature_stats.loc[feature_stats['std'].nlargest(10).index].round(3))

# 5. Advanced Visualization: Learning Curves (for best model)
from sklearn.model_selection import learning_curve

def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None, n_jobs=None, train_sizes=np.linspace(.1, 1.0, 5)):
    plt.figure(figsize=(10, 6))
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    
    plt.grid()
    
    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1, color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g", label="Cross-validation score")
    
    plt.legend(loc="best")
    return plt

# Plot learning curve for age classification (best model)
if 'age' in results and len(results['age']) > 0:
    best_age_model_name = max(results['age'].keys(), key=lambda x: results['age'][x]['accuracy'])
    best_age_model = results['age'][best_age_model_name]['model']
    
    plot_learning_curve(best_age_model, f"Learning Curve - Age Classification\n({best_age_model_name})",
                      X_train['age'], y_train_dict['age'], cv=5)
    plt.show()

print("\n" + "="*50)
print("CONCLUSIONS AND RECOMMENDATIONS")
print("="*50)

print("""
KEY FINDINGS:
1. The dataset contains rich spectral and frequency features for elephant call analysis
2. Multiple classification targets available: age, sex, and behavioral context
3. Ensemble methods (Random Forest, XGBoost) typically perform best for bioacoustic data
4. Feature importance analysis reveals which vocal patterns are most discriminative

RECOMMENDATIONS:
1. Consider collecting more balanced data if any class is underrepresented
2. Experiment with deep learning approaches for potentially better performance
3. Incorporate temporal patterns if sequential data is available
4. Validate findings with domain experts for biological relevance
""")

# Save results
results_summary = {
    'best_models': {},
    'feature_importance': feature_importance.tolist() if 'feature_importance' in locals() else None,
    'accuracy_summary': summary.to_dict()
}

for target_name in results.keys():
    best_model = max(results[target_name].items(), key=lambda x: x[1]['accuracy'])
    results_summary['best_models'][target_name] = {
        'model': best_model[0],
        'accuracy': best_model[1]['accuracy']
    }

print("\nResults summary saved for further analysis.")
```
